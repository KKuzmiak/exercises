###  Data Cleaning 

1.For some Country Codes, there is no corresponding __full Country Name__ available (e.g. for "URS") -> __missing values__ in olympics. Identify these Country Codes and search the Web for the full Country Names. __Replace__ missing values! 

```python
# checking which country abreviations don't have country name 
medals[medals['Country name'].isnull()]['Country code'].unique()
```

```python
# the dictionary with full names
dic2={'ZZX':'Mixed team', 'BOH':'Bohemia', 'ANZ':'Australasia', 'RU1':'Russian Empire', 
      'TCH':'Chechoslovakia', 'YUG':'Yugoslavia', 'ROU':'Romania', 'URS':'Soviet Union', 
      'EUA':'United Team of Germany','BWI':'British West Indies', 'GDR':'East Germany', 
      'FRG':'West Germany', 'EUN':'United Team', 'IOP':'Independent Olpimpic Participants', 
      'SRB':'Serbia', 'TTO':'Trinidad and Tobago', 'MNE':'Montenegro', 'SGP':'Singapore'}
```

```python
# replacing missing values for country names basing on the dictionary above
medals['Country name'] = medals['Country name'].fillna(medals['Country code'].apply(lambda x: dic2.get(x)))
```

```python
# checking results 
medals[medals['Country code']=='ZZX']
```

```python
# checking results 
medals[medals['Country code']=='SRB']
```
