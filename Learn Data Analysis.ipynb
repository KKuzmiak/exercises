### usuwanie niechcianych znaków

```python
Location=(r'/home/kinga/Pulpit/Pandas/learn-data-analysis-w-python-master/datasets/gradedata.csv')
```

```python
df3=pd.read_csv(Location)
```

```python
df3.head()
```

```python
import string
exclude = set(string.punctuation)
def remove_punctuation(x):
    try:
        x = ''.join(ch for ch in x if ch not in exclude)
    except:
        pass
    return x
df3.address = df3.address.apply(remove_punctuation)
df3.head()
```

```python
def remove_whitespace(x):
    try:
        x = ''.join(x.split())
    except:
        pass
    return x
df3.address = df3.address.apply(remove_whitespace)
df3.head()
```

### binning data

```python
Location=(r'/home/kinga/Pulpit/Pandas/learn-data-analysis-w-python-master/datasets/gradedata.csv')
```

```python
df5=pd.read_csv(Location)
```

```python
df5.head()
```

```python
bins = [0, 60, 70, 80, 90, 100]
group_names = ['F', 'D', 'C', 'B', 'A']
df5['lettergrade'] = pd.cut(df5['grade'], bins,labels=group_names)
```

```python
df5.head()
```

```python
?pd.cut
```

```python
bins2=[0,2,4,6,8,10]
group_names2=['I', 'J', 'K', 'L', 'M']
df5['letterhour']=pd.cut(df5['hours'], bins2, labels=group_names2)
df5.head()
```

```python
bins3=[0, 2,4,6] 
group_name3=['s', 'T', 'U',]
df5['letterexercise']=pd.cut(df5['exercise'], bins3, labels=group_name3)
df5.head()
```

```python
bins4=[0, 79.99, 100]
group_name4=['fail', 'pass']
df5['pass_or_fail']=pd.cut(df5['grade'], bins4, labels=group_name4)
df5.head()
```

 ### Applying Functions to Groups, Bins and Columns

```python
## pokazuje ile śrdenio godzin uczyli się uczniowie w poszczególnych grupach
df5.groupby('lettergrade')['hours'].mean()
```

```python
## pokazuje, jaka jest średnia ocena w zależności od płci
gender_preScore = df5['grade'].groupby(df5['gender'])
gender_preScore.mean()
```

```python
## to samo, co wyżej tylko inaczej zapisane
df5.groupby('gender')['grade'].mean()
```

```python
## ile godzin poświęcili na naukę studencji, którzy zdali, a ile ci , którzy nie zdali
df5.groupby('pass_or_fail')['hours'].mean()
```
